```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# إنشاء بيانات وهمية وتحويلها إلى أعداد صحيحة
data = np.array([0, 1, 1, 0, 1, 0, 1, 1], dtype=np.int32)

# تنفيذ عملية XOR بعد تحويل البيانات إلى int
xor_result = np.bitwise_xor(data[:-1], data[1:])

# تحليل الاتجاهات الزمنية باستخدام المتوسط المتحرك
def moving_average(values, window_size):
    return np.convolve(values, np.ones(window_size)/window_size, mode='valid')

ma_3 = moving_average(xor_result, 3)
ma_5 = moving_average(xor_result, 5)

# حساب مصفوفة الارتباط لتحديد العلاقات المخفية
correlation_matrix = np.corrcoef(np.vstack([xor_result[:len(ma_3)], ma_3]))
correlation_strength = correlation_matrix[0, 1]

# تجهيز بيانات الإدخال لـ LSTM
X = np.array([data[i:i+3] for i in range(len(data)-3)], dtype=np.float32).reshape(-1, 1, 3)
y = np.array([data[i+3] for i in range(len(data)-3)], dtype=np.float32)

# بناء نموذج LSTM
model = Sequential([
    LSTM(50, activation='relu', input_shape=(1, 3)),
    Dense(1)
])

# إعداد النموذج للتدريب
model.compile(optimizer='adam', loss='mse')

# تدريب النموذج
model.fit(X, y, epochs=200, verbose=0)

# توقع القيم المستقبلية
future_step = np.array([data[-3:]], dtype=np.float32).reshape(1, 1, 3)
predicted_value = model.predict(future_step)[0][0]

# طباعة النتائج
print("📊 بيانات TensorFlow:", tf.constant(data))
print("🔮 تحليل XOR:", xor_result)
print("📈 متوسط القيم لآخر 3 خطوات:", ma_3)
print("📊 مصفوفة الارتباط:", correlation_matrix)
print("🔬 قوة العلاقة بين القيم:", correlation_strength)
print("🔮 التوقع المستقبلي باستخدام LSTM:", predicted_value)
```
